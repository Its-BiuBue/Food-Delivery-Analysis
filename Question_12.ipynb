```python
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv('final_food_delivery_dataset.csv')
except:
    # Fallback to regeneration if file is missing (robustness)
    orders_df = pd.read_csv('orders.csv')
    users_df = pd.read_json('users.json')
    # No need for restaurants for this specific query, but good for consistency
    merged = pd.merge(orders_df, users_df, on='user_id', how='left')
    df = merged

# Filter for Hyderabad
hyderabad_revenue = df[df['city'] == 'Hyderabad']['total_amount'].sum()

# Round to nearest integer
rounded_revenue = round(hyderabad_revenue)

print(f"Total Revenue from Hyderabad: {hyderabad_revenue}")
print(f"Rounded Revenue: {rounded_revenue}")
```
