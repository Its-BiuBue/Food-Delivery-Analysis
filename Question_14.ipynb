```python
import pandas as pd

# Load the dataset
orders_df = pd.read_csv('orders.csv')
users_df = pd.read_json('users.json')

# Merge orders with users
merged_df = pd.merge(orders_df, users_df, on='user_id', how='left')

# Filter for Gold members
gold_df = merged_df[merged_df['membership'] == 'Gold']

# Calculate average order value
avg_order_value_gold = gold_df['total_amount'].mean()

# Round to 2 decimal places
result = round(avg_order_value_gold, 2)

print(f"Average Order Value for Gold members: {result}")
```
